```python
import pandas as pd

import matplotlib.pyplot as plt
import numpy as np
```

```python
df1 = pd.read_csv("SP5001.csv")
df2 = pd.read_csv("SP5002.csv")
df3 = pd.read_csv("SP5003.csv")
```

```python
print("DF1 : ", df1.shape)
print("DF2 : ", df2.shape)
print("DF3 : ", df3.shape)
```

```python
df = pd.concat([df1,df2,df3],sort=True)
```

```python
df.shape
```

```python
df.loc[:,"Date"] = df.loc[:,"Date"].astype("datetime64")
```

```python
df.sort_values("Date", inplace=True)
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df.drop(["Adj Close", "High", "Low", "Open", "Volume"], axis=1, inplace=True)
```

```python
df.head()
```

```python
df["Year"] = pd.DatetimeIndex(df['Date']).year
```

```python
df["Month"] = pd.DatetimeIndex(df['Date']).month
```

```python
df["Log Close"] = np.log(df["Close"])
```

```python
df["200dayma"] = df["Close"].rolling(200).mean()
```

```python
df.dropna(inplace=True)
```

```python
df["Log 200dayma"] = np.log(df["200dayma"])
```

```python
df["Close"] = df["Close"]
```

```python
df.tail()
```

```python
df.plot("Date", "Log 200dayma", figsize=(20,10))
```

```python
df.plot("Date", "200dayma", figsize=(20,10))
```

```python
df.plot("Date", "Close", figsize=(20,10))
```
